```python
import os
query=input()
urls=[]
files=os.listdir(r'C:\Users\Ramesh kumar\Desktop\memphis notes\IR\Assignment 7\doc2/')
loc = r'C:\Users\Ramesh kumar\Desktop\memphis notes\IR\Assignment 7\doc2/'
#print(files)
all_urls=[]

for file in files:
    
    new1=loc + file
    d= open(new1,'r').read()
    u=d.split('\n',1)[0]
    #print(u)
    all_urls.append(u)
    d1=d
        
    urls.append(d1)
    
```
