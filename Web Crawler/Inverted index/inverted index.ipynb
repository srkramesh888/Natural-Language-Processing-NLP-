```python
import pandas as pd 
import numpy as np
```

```python
import nltk ## importing nltk library for text processing
import re ## importing regular expression
from nltk.corpus import stopwords ## importing stopwords to remove the stopwords
from nltk.stem import WordNetLemmatizer ## importing wordnetlemmatizer to get semantic words
fn=["t1.txt","t2.txt","t3.txt","t5.txt","t6.txt","t7.txt","t8.txt","t9.txt","t10.txt","t11.txt","t12.txt"]
for i in range(len(fn)):
    print(fn[i])
    textname = fn[i]
    text = open(textname,'rt') ## opening file
    t = text.read()
    text.close()
    word=nltk.word_tokenize(t) ## doing tokenization
    word=re.split(r'\W+', t) ## splitting
    corpus=[]
    freq= {}
    lm=WordNetLemmatizer()
    for i in range(0,len(word)):
        if word[i].isalpha(): ## allowing only alphabetical words(removing digits)
            cl=re.sub('[^a-zA-Z]',' ',word[i]) ## removing apart from this
            cl = cl.lower() ## coverting to lower case
            cl = cl.split()
            cl=[lm.lemmatize(word) for word in cl if word not in stopwords.words('english')]
            cl=' '.join(cl)
        if cl in freq:
             freq[cl] = freq[cl]+1
        else:
            freq[cl] = 1
        corpus.append(cl) ## appending all words to corpus
    
        
    
    
    print(len(corpus)) ## printing total count of words in web page
    print(freq) ## finding frequency of a word
    sorted(corpus) ## sorting according to the Alphabetical order
    
```
